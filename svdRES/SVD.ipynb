```python
import numpy as np
import matplotlib.pyplot as plt
import os
 
 
im = plt.imread('/Users/chixinning/Desktop/sample.jpg')
# print('origin_image shape:', im.shape)
if not os.path.exists('/Users/chixinning/Desktop/svdRES'):
	os.mkdir("/Users/chixinning/Desktop/svdRES")
# img = np.zeros(shape=im.shape, dtype='uint8')

```

```python
#分别对1%,2%,10%,50%奇异值的结果
svd_image = []
for ch in range(3):
    im_ch = im[:,:,ch]
    U,D,VT= np.linalg.svd(im_ch)
    imx = np.matmul(np.matmul(U[:,:1],np.diag(D[:1])),VT[:1,:])
    svd_image.append(imx.astype('uint8'))
```

```python
img = np.stack((svd_image[0], svd_image[1], svd_image[2]), 2)
#设置图像保存
plt.figure(figsize=(800/96,1280/96), dpi=96) #设置图像在大小和dpi
plt.imshow(img)
plt.axis('off')
plt.savefig('/Users/chixinning/Desktop/svdRES/1.jpg')
```
